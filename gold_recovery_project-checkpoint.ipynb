```python

def only_similar_columns(data, string):
    new_df = pd.DataFrame()
    namelist = []
    for name in data.columns.tolist():
        if string in name.split('.'):
            new_df = new_df.append(data[name])
            namelist.append(name)
    return new_df, namelist
```

```python
rougher, rougherlist = only_similar_columns(gold_recovery_train, 'rougher')
# rows:
display(rougher)
# columns:
# display(gold_recovery_train[namelist])
```

```python

def find_output_recovery(data, stage):
    # Доля золота в концентрате после флотации/очистки
    c = data['.'.join([stage, 'output', 'concentrate_au'])]
    # Доля золота в сырье/концентрате до флотации/очистки
    f = data['.'.join([stage,'input', 'feed_au'])]
    # Доля золота в отвальных хвостах после флотации/очистки
    t = data['.'.join([stage, 'output', 'tail_au'])]

    output_recovery = pd.Series((c*(f-t))/(f*(c-t))*100)
    return output_recovery
```

```python
my_rougher_output_recovery_train = find_output_recovery(gold_recovery_train, 'rougher')

print("Средняя абсолютная ошибка рассчета эффективности обогащения на этапе флотации:",
        mean_absolute_error(my_rougher_output_recovery_train, gold_recovery_train['rougher.output.recovery']))
print("Число строк с расхождениями в рассчетах:",
      (abs(my_rougher_output_recovery_train - gold_recovery_train['rougher.output.recovery']) > 0).sum())

# display(gold_recovery_train[(abs(my_rougher_output_recovery_train - gold_recovery_train['rougher.output.recovery']) > 0)][rougherlist])
```

# Модель

## Напишем функцию для вычисления итоговой sMAPE

```python
def smape(target, predict):
    smape = sum(abs(target - predict)/((abs(target)+abs(predict))/2))/len(target)*100
    return smape


def total_smape(smape_rougher, smape_final):
    total_smape = 25*smape_rougher/100 + 75*smape_final/100
    return total_smape


smape_scorer = make_scorer(smape, greater_is_better=False)
smape_scorer
```

Теперь есть своя метрика!

## Обучим разные модели и оценим их качество кросс-валидацией. 

### Разделим обучающий датасет на признаки

```python
labels = [item for item in gold_recovery_train.columns.tolist() if item not in gold_recovery_test.columns.tolist()]
labels.append('date')
features_train = gold_recovery_train.drop(labels=labels, axis=1)
target_train = pd.DataFrame(gold_recovery_train[['rougher.output.recovery', 'final.output.recovery']])

features_test = gold_recovery_test.drop(labels='date', axis=1)
target_test = gold_recovery_full[gold_recovery_full['date'].isin(
    [item for item in gold_recovery_test["date"]])][['rougher.output.recovery',
                                                     'final.output.recovery']].reset_index(drop=True)
```

```python
features_train.shape, features_test.shape
```

Привели в соответствие трейн и тест.

### Первый этап обучения (rougher)

```python
# Признаки первого этапа

features_train_rougher_rows, features_train_rougher_list = only_similar_columns(features_train, 'rougher')
features_train_rougher = features_train[features_train_rougher_list]
target_train_rougher = target_train['rougher.output.recovery']

features_test_rougher_rows, features_test_rougher_list = only_similar_columns(features_test, 'rougher')
features_test_rougher = features_test[features_test_rougher_list]
target_test_rougher = target_test['rougher.output.recovery']
```

### Второй этап обучения (final)

```python
# Признаки второго этапа

features_train_final = features_train[[item for item in features_train.columns.tolist() if item not in features_train_rougher.columns.tolist()]]
target_train_final = target_train['final.output.recovery']

features_test_final = features_test[[item for item in features_test.columns.tolist() if item not in features_test_rougher.columns.tolist()]]
target_test_final = target_test['final.output.recovery']
```

### Линейная Регрессия

```python
reg = Pipeline([('transformer', StandardScaler()),
                     ('estimator', LinearRegression())])

smape_linear_regression_rougher = cross_val_score(reg,
                                                  features_train_rougher,
                                                  np.array(target_train_rougher), # передаю сразу массивом
                                                  cv = 5,
                                                  scoring=smape_scorer)
smape_linear_regression_final = cross_val_score(reg,
                                                features_train_final,
                                                np.array(target_train_final),
                                                cv = 5,
                                                scoring=smape_scorer)

print(smape_linear_regression_rougher*(-1), smape_linear_regression_final*(-1))

mean_smape_score_rougher = sum(smape_linear_regression_rougher)/len(smape_linear_regression_rougher)
mean_smape_score_final = sum(smape_linear_regression_final)/len(smape_linear_regression_final)
print('Средняя оценка качества модели для флотации:', mean_smape_score_rougher*(-1))
print('Средняя оценка качества модели для финального концентрата:', mean_smape_score_final*(-1))
print('\n')
print('Итоговое sMAPE:', total_smape(mean_smape_score_rougher*(-1), mean_smape_score_final*(-1)))
```

### Случайный Лес

```python
%%time
seed = 24
cv_k = 5

hyperparams_rf = { 'n_estimators': range (1, 20, 5),
              'max_depth': range (10, 31, 10),
              'min_samples_leaf': range (1,8),
              'min_samples_split': range (2,10,2)
            }

rf = RandomForestRegressor(random_state=seed, criterion='mae')# Логарифм MAE похож на sMAPE
rf_rscv = RandomizedSearchCV(rf,
                             hyperparams_rf, 
                             cv=cv_k, 
                             random_state=seed,
                             scoring=smape_scorer,
                             n_jobs=-1)

rf_rscv.fit(features_train_rougher, target_train_rougher)

rf_rscv_best_params = rf_rscv.best_params_
rf_rscv_best_model_rougher = rf_rscv.best_estimator_
rf_rscv_validation_score_rougher = rf_rscv.best_score_
print('Для флотации:')
print('Лучшие параметры:', rf_rscv_best_params)
print('Лучшая модель:', rf_rscv_best_model_rougher)
print('sMAPE:', rf_rscv_validation_score_rougher*(-1))

rf_rscv.fit(features_train_final, target_train_final)

rf_rscv_best_params = rf_rscv.best_params_
rf_rscv_best_model_final = rf_rscv.best_estimator_
rf_rscv_validation_score_final = rf_rscv.best_score_
print('Для финального концентрата:')
print('Лучшие параметры:', rf_rscv_best_params)
print('Лучшая модель:', rf_rscv_best_model_final)
print('sMAPE:', rf_rscv_validation_score_final*(-1))

print('\n')
print('Итоговое sMAPE:', total_smape(rf_rscv_validation_score_rougher*(-1), rf_rscv_validation_score_final*(-1)))
```

## Проверим лучшую модель на тестовой выборке

```python
train_rougher = pd.concat([features_train_rougher, target_train_rougher], axis=1)
sample_train_rougher = train_rougher.sample(n=len(features_test_rougher))
features_sample_train_rougher = sample_train_rougher.drop(labels='rougher.output.recovery', axis=1)
target_sample_train_rougher = sample_train_rougher['rougher.output.recovery']

train_final = pd.concat([features_train_final, target_train_final], axis=1)
sample_train_final = train_final.sample(n=len(features_test_final))
features_sample_train_final = sample_train_final.drop(labels='final.output.recovery', axis=1)
target_sample_train_final = sample_train_final['final.output.recovery']
```

```python
rf_test = RandomForestRegressor(n_estimators=16,
                                min_samples_split=2,
                                min_samples_leaf=7,
                                max_depth=20,
                                random_state=123,
                                criterion='mae')
rf_test.fit(features_sample_train_rougher, target_sample_train_rougher)
rf_predict_test_rougher = rf_test.predict(features_test_rougher)
smape_rougher_test = smape(target_test_rougher, rf_predict_test_rougher)
print(smape_rougher_test)

rf_test.fit(features_sample_train_final, target_sample_train_final)
rf_predict_test_final = rf_test.predict(features_test_final)
smape_final_test = smape(target_test_final, rf_predict_test_final)
print(smape_final_test)

print('\n')
print(total_smape(smape_rougher_test, smape_final_test))
```
